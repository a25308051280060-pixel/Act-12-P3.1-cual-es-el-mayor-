## las variables
1. A --> primer valor para comprar
2. B --> segundo valor para comprar
3. M --> resultado de la comparacion

## formula
* M = A
* M = B
* A > B

```python
# calcula la diferencia de valor if else
# datos de entrada A, B
# Mostrar B si es falso
# mostrar A si es verdadero
A = int(input("ingrese el primer valor"))
B = int(input("ingres el segundo valor"))
if A > B :
   M = A
else:
   M = B
print("el valor mayor es:", M)
print("fin del programa")
print(" programa realizado por Ambar Fierro NL = 19")
```
